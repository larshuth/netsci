```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
p = 0.2
q = 1 - p

max_n = int(5 / p)
ns = np.arange(0, max_n)

probs = q ** ns * p
cumprobs = probs.cumsum() 

stepx = np.zeros((1+2*max_n, ))
stepx[1::2] = ns
stepx[2:-1:2] = ns[1:]
stepx[-1] = max_n

stepy = np.empty((1+2*max_n, ))
stepy[1::2] = cumprobs
stepy[2::2] = cumprobs

plt.plot(ns, probs, label="Wahrscheinlichkeit $f(x)$", marker="x")
plt.plot(stepx, stepy, label="Verteilung $F(x)$")
plt.xlabel("Anzahl Misserfolge $X$ vor erstem Erfolg")
plt.ylabel("Wahrscheinlichkeit")
plt.title("Geometrische Verteilung mit $p= %.1f$" % p)
plt.legend()

plt.savefig("geometric-distr.pdf", bbox_inches="tight")
```

```python
plt.plot(stepy, stepx, color="C1")
plt.ylabel("Anzahl Misserfolge $X$ vor erstem Erfolg")
plt.xlabel("Wahrscheinlichkeit")
plt.title("Umkehrfunktion $F^{-1}(p)$")
plt.legend()

plt.savefig("geometric-distr-inv.pdf", bbox_inches="tight")
```

```python
Us = np.arange(1e-9, 1, step = 0.005)
invers = np.log(1 - Us) / np.log(0.8) - 1

plt.plot(stepy, stepx, color="C1", label=r"$\lceil \hat F^{-1}(U) \rceil$")
plt.plot(Us, invers, color="C2", label=r"$\hat F^{-1}(U)$")
plt.xlabel("Eingabe $U$")
plt.xlabel("Funktionswert")
plt.legend()

plt.savefig("geometric-distr-inv_hat.pdf", bbox_inches="tight")
```
